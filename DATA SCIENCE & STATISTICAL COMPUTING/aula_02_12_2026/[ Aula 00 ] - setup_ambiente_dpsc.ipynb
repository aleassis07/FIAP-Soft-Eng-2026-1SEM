## 1) Windows — criar o ambiente (Anaconda Prompt)

### 1.1 abrir o Anaconda Prompt
- menu iniciar → procure por **Anaconda Prompt**

### 1.2 ir até a pasta do curso
Substitua o caminho pelo seu.

```bash
cd C:\Users\SEU_USUARIO\Downloads\DPSC_Aula01
```

### 1.3 criar o ambiente a partir do arquivo `yml`

```bash
conda env create -f environment_dpsc_aula01.yml
```

### 1.4 ativar o ambiente

```bash
conda activate dpsc_aula01
```

### 1.5 registrar o kernel no Jupyter

```bash
python -m ipykernel install --user --name dpsc_aula01 --display-name "Python (dpsc_aula01)"
```

### 1.6 abrir o JupyterLab

```bash
jupyter lab
```

## 4) validação do ambiente (rode as células abaixo)

Agora sim: execute as células desta seção para confirmar que está tudo instalado.

✅ se não aparecer erro de importação, você está pronto.

````python
import sys
import numpy as np
import pandas as pd
import sklearn
import matplotlib

print("Python:", sys.version.split()[0])
print("numpy:", np.__version__)
print("pandas:", pd.__version__)
print("scikit-learn:", sklearn.__version__)
print("matplotlib:", matplotlib.__version__)
print("executável:", sys.executable)
````

### teste rápido com o dataset Wine

````python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt

wine = load_wine(as_frame=True)
df = wine.frame

print("shape:", df.shape)
print("contagem por classe:", df["target"].value_counts().sort_index().to_dict())

counts = df["target"].value_counts().sort_index()

plt.figure()
plt.bar(counts.index.astype(str), counts.values)
plt.xlabel("classe (target)")
plt.ylabel("contagem")
plt.title("Wine: contagem por classe")
plt.show()
````

## 5) solução de problemas (FAQ)

### 5.1 erro: `ModuleNotFoundError: No module named ...`
Isso quase sempre significa **kernel errado**.

Faça:
1. **Kernel → Change Kernel → Python (dpsc_aula01)**
2. rode novamente a célula

Se persistir, no Terminal/Prompt:
```bash
conda activate dpsc_aula01
conda install -y numpy pandas matplotlib scikit-learn jupyterlab ipykernel
```

---

### 5.2 kernel `Python (dpsc_aula01)` não aparece
No Terminal/Prompt:
```bash
conda activate dpsc_aula01
python -m ipykernel install --user --name dpsc_aula01 --display-name "Python (dpsc_aula01)"
```
Feche e reabra o JupyterLab.

---

### 5.3 erro ao criar o ambiente (`conda env create ...`)
- confira se você está **na pasta correta** (onde está o `.yml`)
- confira se o arquivo se chama exatamente `environment_dpsc_aula01.yml`

Você pode listar os arquivos da pasta:
- Windows: `dir`
- mac/Linux: `ls`
